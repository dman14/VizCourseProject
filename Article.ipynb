## Europe's Economy

**Economical Overview of Europe**

To have a better overlook over the Europe's economy, IMF supplied quite a lot of data with everything related to it, such as gross domestic product, or GDP, GDP per capita, inflation percentages, unemployment rate, and so on. Included into this, besides the actual data until 2019, IMF provided their estimated values too for the end of 2020 and 2021, and updated this April.

**What can we see from this?**

Of course looking directly at all the data doesn't give us a lot of information, but analyzing it a bit further, we can see some details which are going to be addressed.

Firstly, even if it doesn't look that good for 2020, and overall having a loss in GDP by 6.611%, the predictions are that Europe's going to recover quite well after 2021, having an increase in GDP of 4.986%.

**The 2008**

This can be further compared with another crisis that happened not long ago, namly the Financial crisis of 2008, and at least in predictions, even though the fall was bigger as a lot of businesses had to shutdown, the estimate recovery is way faster then the one from 2008, presumembly not affecting the economy as bad as before.

**The differences in Europe**

Economically speaking, Europe can be split into two parts, the Euro area, and Emerging and developing Europe. This split can be seen on IMF too, we're the data set of Euro zone contains bigger countries like Germany, Spain, Italy, etc. while in the Emerging and developing economies we can find countries such as Albania, Montenegro, Belarus, Poland, etc.


Interestingly enough, we can spot those countries on the map even by just plotting the percentage growth of the GDP per capita, as even the name implies(emerging and developing economies) the economy of such countries grow at a significantly higher rate than other fully developed ones.

This fact about the Developing economy countries in Europe can be seen in those times of economic downfall, as those countries, on average, did not get such a harsh fall as the Euro zone countries, and, presumably, will advance quite a bit faster after in 2021 too.
